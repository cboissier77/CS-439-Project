# Model 1

```python
# Load the data from pickle files
with open('pickle_lists/curri_losses_1_list.pkl', 'rb') as f:
    curri_losses_1_list = pickle.load(f)

with open('pickle_lists/times_curri_1_list.pkl', 'rb') as f:
    times_curri_1_list = pickle.load(f)

with open('pickle_lists/acc_curri_1_list.pkl', 'rb') as f:
    acc_curri_1_list = pickle.load(f)
```

```python
# Load the data from pickle files

with open('pickle_lists/anti_curri_losses_1_list.pkl', 'rb') as f:
    anti_curri_losses_1_list = pickle.load(f)

with open('pickle_lists/times_anti_curri_1_list.pkl', 'rb') as f:
    times_anti_curri_1_list = pickle.load(f)

with open('pickle_lists/acc_anti_curri_1_list.pkl', 'rb') as f:
    acc_anti_curri_1_list = pickle.load(f)
```

```python
# Load the data from pickle files

with open('pickle_lists/rand_losses_1_list.pkl', 'rb') as f:
    rand_losses_1_list = pickle.load(f)

with open('pickle_lists/times_rand_1_list.pkl', 'rb') as f:
    times_rand_1_list = pickle.load(f)

with open('pickle_lists/acc_rand_1_list.pkl', 'rb') as f:
    acc_rand_1_list = pickle.load(f)
```

```python
# Load the data from pickle files
with open('pickle_lists/std_curri_losses_1_list.pkl', 'rb') as f:
    std_curri_losses_1_list = pickle.load(f)

with open('pickle_lists/times_std_curri_1_list.pkl', 'rb') as f:
    times_std_curri_1_list = pickle.load(f)

with open('pickle_lists/acc_std_curri_1_list.pkl', 'rb') as f:
    acc_std_curri_1_list = pickle.load(f)
```

```python
data = pd.DataFrame({
    'Value': acc_std_curri_1_list + acc_curri_1_list + acc_rand_1_list + acc_anti_curri_1_list,
    'Method': ['Standard curriculum']*10 + ['Curriculum'] * 10 + ['Random'] * 10 + ['Anti-Curriculum'] * 10
})
plt.figure(figsize=(6, 5))
sns.barplot(x='Method', y='Value', data=data, ci=None, estimator='mean', palette='Set2')
values = [np.mean(acc_std_curri_1_list) , np.mean(acc_curri_1_list),  np.mean(acc_rand_1_list) , np.mean(acc_anti_curri_1_list)]
for i, v in enumerate(values):
    plt.text(i, v + 0.02, f'{v:.4f}', ha='center', va='bottom', fontsize=10 )
sns.stripplot(x='Method', y='Value', data=data, color='black', alpha=0.7 , size = 8)
plt.ylim(min(values) - 0.1, max(values) + 0.1)
plt.title("accuracy of model 1")
plt.ylabel("Accuracy value")
plt.tight_layout()
plt.show()
```

```python
data = pd.DataFrame({
    'Value': times_std_curri_1_list + times_curri_1_list + times_rand_1_list + times_anti_curri_1_list,
    'Method': ['Standard Curriculum'] * 10+ ['Curriculum'] * 10 + ['Random'] * 10 + ['Anti-Curriculum'] * 10
})
plt.figure(figsize=(6, 5))
sns.barplot(x='Method', y='Value', data=data, ci=None, estimator='mean', palette='Set2')
values = [np.mean(times_std_curri_1_list),np.mean(times_curri_1_list), np.mean(times_rand_1_list), np.mean(times_anti_curri_1_list)]
for i, v in enumerate(values):
    plt.text(i, v + 5, f'{v:.4f}', ha='left', va='bottom', fontsize=10 )
sns.stripplot(x='Method', y='Value', data=data, color='black', alpha=0.7 , size = 8)
plt.ylim(min(values) - 30, max(values) + 50)
plt.title("Time of execution of model 1")
plt.ylabel("Execution time")
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))  

# -------- Curriculum --------
for i in range(10):
    axes[0].plot(curri_losses_1_list[i], label=f'iter {i+1}')

axes[0].set_title('Curriculum')
axes[0].set_xlabel('Step')
axes[0].legend()
axes[0].set_ylabel('Loss')
axes[0].grid(True)

# -------- Anti-curriculum --------
for i in range (10):
    axes[1].plot(anti_curri_losses_1_list[i], label=f'iter {i+1}')

axes[1].set_title('Anti_curriculum')
axes[1].set_xlabel('Step')
axes[1].set_ylabel('Loss')
axes[1].legend()
axes[1].grid(True)

# --------  Random --------
for i in range (10):
    axes[2].plot(rand_losses_1_list[i], label=f'iter {i+1}')
axes[2].set_title('Random')
axes[2].set_xlabel('Step')
axes[2].set_ylabel('Loss')
axes[2].grid(True)
axes[2].legend()
plt.tight_layout()
plt.show()
```
